```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('CAC.csv')
```

```python
wifi_satisfaction = df.groupby('Course')[' Wi-Fi speed and reliability?'].value_counts().unstack().fillna(0)
wifi_satisfaction.plot(kind='bar')
plt.title('Satisfaction with Wifi Services on Campus')
plt.xlabel('Satisfaction Level')
plt.ylabel('Count')
plt.show()
```

The above graph is intended to analyze and visualize satisfaction with Wi-Fi services on campus by course. It creates a bar chart showing the count of each satisfaction level within each course

```python
healthcare_satisfaction = df.groupby('Gender')['Healthcare services on campus?'].value_counts()
healthcare_satisfaction.plot(kind='bar')
plt.title('Satisfaction with Healthcare Services on Campus')
plt.xlabel('Satisfaction Level')
plt.ylabel('Count')
plt.show()
```

The code groups the data by gender and the satisfaction level with healthcare services on campus. It then creates a bar chart to display the count of individuals in each satisfaction level for each gender. This can provide insights into how different genders perceive healthcare services on campus.

```python
course_satisfaction_grouped = df.groupby('Course')['Satisfaction with campus facilities'].mean()
course_satisfaction_grouped.plot(kind='bar')
plt.title('Average Satisfaction by Course')
plt.xlabel('Course')
plt.ylabel('Average Satisfaction')
plt.xticks(rotation=45)
plt.show()
```

The code calculates the mean satisfaction level with campus facilities for each course and creates a bar chart to display these average satisfaction levels. This chart can help identify which courses tend to have higher or lower satisfaction with campus facilities.

```python
# Group by Gender and analyze Wi-Fi speed
gender_wifi_grouped = df.groupby('Course')['Gender'].value_counts()

# Create a histogram
gender_wifi_grouped.plot(kind='bar', edgecolor='black')
plt.title('COURSE VS GENDER')
plt.xlabel('Course')
plt.ylabel('Gender')
plt.show()

```

It explores the distribution of gender within different courses.

```python
# e. Environmental Responsibility Perception
environmental_responsibility = df['Do you think the campus is taking adequate measures to promote sustainability and environmental responsibility in its facilities (e.g., recycling, energy efficiency)?'].value_counts()
environmental_responsibility.plot(kind='bar')
plt.title('Environmental Responsibility Perception')
plt.xlabel('Perception')
plt.ylabel('Count')
plt.show()
```

The code counts the responses related to the perception of the campus taking measures for sustainability and environmental responsibility. It then creates a bar chart to display the count of different perceptions, which can give insights into how the campus community views environmental responsibility efforts.

```python
# d. Satisfaction with Campus Facilities
df['Satisfaction with campus facilities'].hist(bins=5, edgecolor='black')
plt.title('Satisfaction with Campus Facilities')
plt.xlabel('Satisfaction Level')
plt.ylabel('Count')
plt.show()
```

The code generates a histogram showing the distribution of satisfaction levels with campus facilities. This visualization can help identify the most common satisfaction levels and whether there are any extreme outliers.

```python
# Calculate the course distribution
course_distribution = df['Course'].value_counts()

# Create a bar graph
plt.figure(figsize=(10, 6))
sns.barplot(x=course_distribution.index, y=course_distribution)
plt.title('Course Distribution')
plt.xlabel('Course')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

The code creates a bar chart to display the distribution of different courses among the dataset. This can provide insights into the popularity of various courses within the surveyed population

```python
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

The first pie chart displays the distribution of gender in the dataset. It provides a visual representation of the proportion of each gender in the survey.

```python
category_counts = df['Transportation facilities in Lavasa?'].value_counts()
category_counts.plot(kind='bar')
plt.title('Transportation Facilities Satisfaction')
plt.xlabel('Satisfaction Level')
plt.ylabel('Count')
```

The code creates a bar chart that shows the count of individuals at different satisfaction levels with transportation facilities in Lavasa. This visualization helps in understanding how satisfied or dissatisfied respondents are with transportation options.

```python
healthcare_satisfaction = df['Healthcare services on campus?'].value_counts()
plt.pie(healthcare_satisfaction, labels=healthcare_satisfaction.index, autopct='%1.1f%%')
plt.title('Healthcare Services Satisfaction')
plt.axis('equal')
```

The second pie chart displays the distribution of satisfaction levels with healthcare services on campus. It provides a visual representation of the proportion of respondents in each satisfaction level.

```python
df_grouped = df.groupby(['Gender', 'Course'])['Transportation facilities in Lavasa?'].value_counts().unstack().fillna(0)
df_grouped.plot(kind='bar', stacked=True)
plt.title('Transportation Facilities Satisfaction by Gender and Course')
```

The code groups the data by gender, course, and satisfaction levels with transportation facilities. It creates a stacked bar chart to show how satisfaction with transportation facilities varies across different genders and courses. This can help identify patterns and differences in satisfaction.

```python
cross_tab = pd.crosstab(df['Gender'], df['Transportation facilities in Lavasa?'])
sns.heatmap(cross_tab, annot=True, cmap='coolwarm', fmt='d')
plt.title('Transportation Facilities Satisfaction by Gender')
```

The heatmap shows the satisfaction with transportation facilities by gender. It provides a clear visual representation of how different genders perceive transportation facilities, with annotated counts for each combination of gender and satisfaction level.
